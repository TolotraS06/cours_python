## MATPLOTLIB

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = {f"experience{i}":np.random.randn(100) for i in range (4)}
dataset
```

```python
def  graphique (data):
    taille = len(data)
    y = plt.figure(figsize=(10,9))
    j= 1;
    for i in dataset:
        X =[k for k in range (0,len(data[i]))]
        plt.subplot (taille,1,j)
        plt.plot(X,data[i])
        plt.title(i)
        j+=1
    plt.show()
      
graphique(dataset)    
```
